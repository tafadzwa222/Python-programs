```python
#Example C2-The Grading Weight System
# Program: Shop Product Discount System

import numpy as np

# Products sold in a retail shop with their original prices.
# Initially, every product has a multiplier of 1.0 (full price).

product_names = np.array([
    "Laptop", "Headphones", "T-Shirt", "Jeans",
    "Rice Bag", "Coffee", "Smartphone", "Jacket"
])

# Original prices in Indian Rupees
original_prices = np.array([
    50000, 3000, 1200, 2500,
    900, 400, 30000, 3500
])

# All products start at full price
price_multipliers = np.ones(8)

# Apply category-based discount multipliers:
# Electronics = 20% discount  -> multiplier 0.80
# Clothing = 30% discount     -> multiplier 0.70
# Food = 10% discount         -> multiplier 0.90

price_multipliers = np.array([
    0.80, 0.80, 0.70, 0.70,
    0.90, 0.90, 0.80, 0.70
])

# Calculate discounted prices
discounted_prices = original_prices * price_multipliers

# Question (i):
# Compare total revenue at full price and after discounts.

full_price_revenue = np.sum(original_prices)

discounted_revenue = np.sum(discounted_prices)

print("Question (i)")
print("Total revenue at full price: ₹", full_price_revenue)
print("Total revenue after discounts: ₹", discounted_revenue)

# Question (ii):
# Find which product gives the best deal in absolute terms.
# Best deal = maximum amount saved.

savings = original_prices - discounted_prices

best_deal_index = np.argmax(savings)

print("\nQuestion (ii)")
print("Product offering the best deal:", product_names[best_deal_index])
print("Amount saved: ₹", savings[best_deal_index])

# Question (iii):
# Calculate percentage revenue lost due to discounts.

revenue_loss = full_price_revenue - discounted_revenue

percentage_loss = (revenue_loss / full_price_revenue) * 100

print("\nQuestion (iii)")
print("Percentage of revenue lost due to discounts:",
      percentage_loss, "%")
```

```python
#The Perfomance Multiplier
# Program: Employee Performance Multiplier System

import numpy as np

# Employees in an organisation start with a base performance value
# of 1.0. Each employee receives a multiplier based on performance.
# Multiplier > 1.0 means above expected performance.
# Multiplier < 1.0 means below expected performance.

employee_names = np.array([
    "Employee A", "Employee B", "Employee C",
    "Employee D", "Employee E", "Employee F",
    "Employee G", "Employee H"
])

# Everyone starts with a base value of 1.0
base_value = np.ones(8)

# Personalised performance multipliers based on work output
performance_multipliers = np.array([
    1.15, 0.95, 1.25, 1.05,
    0.85, 1.10, 1.30, 0.90
])

# Final performance values after applying multipliers
final_performance = base_value * performance_multipliers


# Question (i):
# What is the final range between the highest and lowest performer?

performance_range = np.max(final_performance) - np.min(final_performance)

highest_index = np.argmax(final_performance)
lowest_index = np.argmin(final_performance)

print("Question (i)")
print("Highest performer:", employee_names[highest_index],
      "with multiplier", final_performance[highest_index])

print("Lowest performer:", employee_names[lowest_index],
      "with multiplier", final_performance[lowest_index])

print("Range between highest and lowest performer:",
      performance_range)

# Question (ii):
# What is the average multiplier applied across the group?

average_multiplier = np.mean(performance_multipliers)

print("\nQuestion (ii)")
print("Average multiplier across the group:",
      average_multiplier)

# Question (iii):
# How many performers are above the group average after multipliers?

performers_above_average = np.sum(
    final_performance > average_multiplier
)

print("\nQuestion (iii)")
print("Number of performers above group average:",
      performers_above_average)
```

```python
#Example D3-The What-If Explorer

import numpy as np

# Real-world scenario:
# A manufacturing company wants to estimate the probability of a product
# being defective using random simulation.
#
# 1 = Defective product
# 0 = Good product

number_of_products = 100000

np.random.seed(1)

simulation_one = np.random.choice(
    [0, 1],
    size=number_of_products,
    p=[0.97, 0.03]
)

defective_probability_one = np.mean(simulation_one)


# Second simulation using seed 2

np.random.seed(2)

simulation_two = np.random.choice(
    [0, 1],
    size=number_of_products,
    p=[0.97, 0.03]
)

defective_probability_two = np.mean(simulation_two)

# Question (i):
# What did the simulation predict for the probability of defects?


print("Question (i)")
print("Predicted defect probability in Simulation 1:",
      round(defective_probability_one * 100, 2), "%")

# Question (ii):
# Compare results from two simulations with different seeds.

variation = abs(defective_probability_one - defective_probability_two)

print("\nQuestion (ii)")
print("Predicted defect probability in Simulation 2:",
      round(defective_probability_two * 100, 2), "%")

print("Variation between two simulations:",
      round(variation * 100, 4), "%")

# Question (iii):
# Conclusion of the Monte Carlo simulation.

print("\nQuestion (iii)")
print("Conclusion: Large random simulations provide a stable estimate of defect probability.")
```

```python
#Example F3-The Pattern Discovery
# Program: Trace and Determinant Pattern of Identity Matrices

import numpy as np

# Create identity matrices of sizes 2, 3, 4, and 5.

identity_2 = np.eye(2)
identity_3 = np.eye(3)
identity_4 = np.eye(4)
identity_5 = np.eye(5)

identity_matrices = [
    ("2x2 Identity Matrix", identity_2),
    ("3x3 Identity Matrix", identity_3),
    ("4x4 Identity Matrix", identity_4),
    ("5x5 Identity Matrix", identity_5)
]

# Calculate trace and determinant for each identity matrix.

for name, matrix in identity_matrices:
    trace_value = np.trace(matrix)
    determinant_value = np.linalg.det(matrix)

    print(name)
    print("Trace:", trace_value)
    print("Determinant:", determinant_value)
    print("-----------")

# Question (i):
# Pattern: The trace of an identity matrix is equal to its size because
# all diagonal elements are 1 and there are n diagonal elements.

# Question (ii):
# Pattern: The determinant of every identity matrix is always 1,
# regardless of its size.

# Question (iii):
# General rule:
# For an n x n identity matrix, trace(I) = n and determinant(I) = 1.
```

```python
#Example G1-The Scale Builder
# Program: Student Stress Evaluation Scale

import numpy as np

# Real-world scenario:
# A stress evaluation scale is created from 0 to 100.
# 0 = No stress and 100 = Extremely high stress.
# linspace is used because we want fixed divisions on the scale.

stress_scale = np.linspace(0, 100, 11)

print("Stress Evaluation Scale:")
print(stress_scale)

# Realistic stress measurements collected from 5 students.
# Values represent stress scores out of 100.

student_stress_scores = np.array([
    25, 45, 60, 78, 90
])

# Define zones:
# Low stress    : 0 - 39
# Medium stress : 40 - 69
# High stress   : 70 - 100

low_stress = student_stress_scores[
    student_stress_scores < 40
]

medium_stress = student_stress_scores[
    (student_stress_scores >= 40) &
    (student_stress_scores < 70)
]

high_stress = student_stress_scores[
    student_stress_scores >= 70
]


# Question (i):
# Apply measurements to the evaluation scale.

print("\nStudent stress measurements:")
print(student_stress_scores)

# Question (ii):
# Classify each measurement into its defined zone.

print("\nClassification of measurements:")

print("Low stress zone:", low_stress)
print("Medium stress zone:", medium_stress)
print("High stress zone:", high_stress)

# Question (iii):
# Calculate percentage of measurements in each zone.

total_measurements = len(student_stress_scores)

low_percentage = (len(low_stress) / total_measurements) * 100
medium_percentage = (len(medium_stress) / total_measurements) * 100
high_percentage = (len(high_stress) / total_measurements) * 100

print("\nPercentage in each zone:")
print("Low stress:", low_percentage, "%")
print("Medium stress:", medium_percentage, "%")
print("High stress:", high_percentage, "%")
```

```python
#Example G2-The Schedule Generator
# Program: Clinic Appointment Time Slot Scheduler

import numpy as np
# Real-world scenario:
# A clinic schedules appointments every 30 minutes from 9:00 AM to 5:00 PM.
# Each generated value represents an available appointment slot.

# Generate time slots in minutes from 9:00 AM (540 minutes)
# to 5:00 PM (1020 minutes) with 30-minute intervals.

appointment_minutes = np.arange(540, 1020, 30)

# Convert minutes into readable time format

appointment_times = []

for minutes in appointment_minutes:
    hour = minutes // 60
    minute = minutes % 60
    period = "AM" if hour < 12 else "PM"

    display_hour = hour if hour <= 12 else hour - 12

    appointment_times.append(
        f"{display_hour}:{minute:02d} {period}"
    )

appointment_times = np.array(appointment_times)

# Question (i):
# How many total slots are generated?

total_slots = len(appointment_times)

print("Question (i)")
print("Total appointment slots generated:",
      total_slots)

# Question (ii):
# Which slot number is closest to noon?

noon_minutes = 720  # 12:00 PM

closest_slot_index = np.argmin(
    abs(appointment_minutes - noon_minutes)
)

print("\nQuestion (ii)")
print("Slot closest to noon:")
print("Slot number:", closest_slot_index + 1)
print("Time:", appointment_times[closest_slot_index])

# Question (iii):
# If 3 slots are cancelled, find remaining slots and utilisation rate.

cancelled_slots = 3

remaining_slots = total_slots - cancelled_slots

utilisation_rate = (
    remaining_slots / total_slots
) * 100

print("\nQuestion (iii)")
print("Slots remaining after cancellation:",
      remaining_slots)

print("New utilisation rate:",
      round(utilisation_rate, 2), "%")
```

```python
#Example G3-The Discovery
# Program: Travel Time Savings Threshold Analysis

import numpy as np
# Scenario:
# Compare travel time at two speeds:
# Route 1 speed = 40 km/hour
# Route 2 speed = 60 km/hour
# Find the distance where the faster speed saves more than 1 hour.
#
# np.arange() is used because we need a fixed step size of 10 km
# between each distance value.

distances = np.arange(0, 201, 10)

# Calculate travel times in hours
time_at_40_kmph = distances / 40
time_at_60_kmph = distances / 60

# Calculate time saved by travelling at 60 kmph
time_saved = time_at_40_kmph - time_at_60_kmph

# Question (i):
# Identify the exact threshold distance where saving exceeds 1 hour.
threshold_index = np.argmax(time_saved > 1)

threshold_distance = distances[threshold_index]

print("Question (i)")
print("Distance where travel time saving exceeds 1 hour:",
      threshold_distance, "km")

print("Time saved at this distance:",
      round(time_saved[threshold_index], 2), "hours")

# Question (ii):
# One-line conclusion.

# Conclusion: Travelling at 60 kmph saves more than 1 hour compared to 40 kmph after crossing the 120 km distance.

# Question (iii):
# Function used and reason.

# np.arange() was used because the distance values require a fixed interval of 10 km.
```
