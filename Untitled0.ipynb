```python
# Program: Comparing monthly electricity consumption of two households
import numpy as np

# Monthly electricity consumption (in kilowatt-hours, kWh)
# for two households over 8 months.

house_a_consumption = np.array([325, 340, 318, 360, 355, 370, 348, 362])

house_b_consumption = np.array([310, 335, 325, 345, 342, 360, 338, 350])

# Find the average monthly electricity consumption of each house
# and calculate the difference between the averages.

average_house_a = np.mean(house_a_consumption)
average_house_b = np.mean(house_b_consumption)

average_difference = abs(average_house_a - average_house_b)

print("Question (i)")
print("Average consumption of House A:", average_house_a, "kWh")
print("Average consumption of House B:", average_house_b, "kWh")
print("Difference between the averages:", average_difference, "kWh")

# Question (ii):
# Count the number of months in which House A consumed more
# electricity than House B.

months_house_a_higher = np.sum(house_a_consumption > house_b_consumption)

print("\nQuestion (ii)")
print("House A consumed more electricity than House B in",
      months_house_a_higher, "months.")

# Question (iii):
# Find the gap between the maximum electricity consumption values.

maximum_gap = abs(np.max(house_a_consumption) - np.max(house_b_consumption))

print("\nQuestion (iii)")
print("Gap between the maximum consumption values:", maximum_gap, "kWh")
```
